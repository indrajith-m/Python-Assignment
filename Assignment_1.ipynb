**Question 3. Compare and contrast mutable and immutable objects in python with example.**


Answer:

In Python, objects (values stored in variables) can be categorized into mutable and immutable types based on whether their state or value can be changed after they are created. Understanding the difference between these two is crucial for writing efficient and bug-free code.

Mutable Objects:

Mutable objects are those whose values or contents can be changed after they are created. This means that you can modify the object in place without creating a new object.

Characteristics of Mutable Objects:

Their contents can be changed after creation.
Changes to the object affect all references pointing to the same object.
They are generally more efficient in terms of memory usage since modifications do not require the creation of a new object.

Common Mutable Types:

* Lists
* Dictionaries
* Sets
* User-defined classes (unless explicitly made immutable)

Example:

Creating a mutable object (list)

my_list = [1, 2, 3]

Modifying the list (changing the value of an element)

my_list[1] = 4  # Changes the second element from 2 to 4

print(my_list)

Output: [1, 4, 3]

Adding an element to the list

my_list.append(5)

print(my_list)

Output: [1, 4, 3, 5]



Immutable Objects:

Immutable objects, on the other hand, cannot have their values or contents changed once they are created. Any operation that tries to modify the object will result in the creation of a new object.

Characteristics of Immutable Objects:

Their contents cannot be changed after they are created.
Any modification results in the creation of a new object.
They are typically safer in a multi-threaded environment because their state cannot be altered accidentally.

Common Immutable Types:
* Integers
* Floats
* Strings
* Tuples
* Frozensets
* Namedtuples

Example:

Creating an immutable object (string)

name = "Alice"

Modifying the string creates a new string object

new_name = name.replace("Alice", "Bob")  # A new string object is created

print(name)

Output: Alice (original string is unchanged)

print(new_name)

Output: Bob (new string object)

* Right shift : >>

a = 5  # 0101 in binary

print(a ^ b)   Output: 6 (0101 ^ 0011 = 0110)

print(a >> 1)  Output: 2  (right shift: 0101 >> 1 = 0010)



**Question 5. Explain the concept of type casting in python with example.**

Answer:

Type Casting in Python:

Type casting (or type conversion) in Python refers to the process of converting one data type into another. Python has two main types of type casting:

* Implicit Type Casting (Automatic Type Conversion)
* Explicit Type Casting (Manual Type Conversion)


type-conversion-in-python.webp





1. Implicit Type Casting (Automatic Type Conversion)

In implicit type casting, Python automatically converts one data type to another without the user having to specify the type. This usually happens when a smaller data type is converted to a larger data type, to prevent data loss. Python does this implicitly when it determines that the conversion will not result in a loss of data or precision.

For example:

* Converting an integer to a float

* Converting a smaller integer to a larger integer

* Combining different types in an expression

Examples:

Integer to float (automatic)

x = 5     # integer

y = 2.5   # float


result = x + y  # Python automatically converts x to a float before adding

print(result)  # Output: 7.5 (float)




2. Explicit Type Casting (Manual Type Conversion)

Explicit type casting is when the programmer manually converts one data type to another using Python's built-in functions. This is done using functions such as int(), float(), str(), list(), etc.


Common Type Conversion Functions:

* Converts a value to an integer : int()

* Converts a value to a float : float()

* Converts a value to a string : str()

* Converts an iterable (like a tuple, string, etc.) to a list : list()

* Converts an iterable to a tuple : tuple()

* Converts an iterable to a set : set()

Examples:

1. Converting a string to an integer:

num_str = "100"

num_int = int(num_str)  # Convert string to integer

print(num_int)  # Output: 100 (integer)


2. Converting an integer to a string:

num = 123

num_str = str(num)  # Convert integer to string

print(num_str)  # Output: "123" (string)

3. Converting a float to an integer:

pi = 3.14159

pi_int = int(pi)  # Convert float to integer (this truncates the decimal
part)

print(pi_int)  # Output: 3 (integer)




**Question 6. How do Conditional statements work in python? Illustrate with example.**

Answer:

Conditional statements in Python are used to execute certain blocks of code based on whether a condition (expression) is True or False. They allow you to control the flow of the program by choosing different paths of execution based on logical conditions.


Python has several types of conditional statements:

* if Statement

* if-else Statement

* if-elif-else Statement

* Nested Conditional Statements


Conditional-Statements-450x338.jpg








1. The if Statement

The simplest form of a conditional statement. It runs a block of code only if a specified condition is True.

Example:

x = 7

if x > 5:
    
print("x is greater than 5")

output: x is greater than 5


2. The if-else Statement

The if-else statement allows for an alternative action if the condition evaluates to False. If the condition is True, the code inside the if block is executed; otherwise, the code inside the else block is executed.

Example:

x = 3

if x > 5:

print("x is greater than 5")

else:

print("x is not greater than 5")


output:  x is not greater than 5


3. The if-elif-else Statement

The if-elif-else statement allows you to check multiple conditions. If the first if condition is False, it checks the next elif condition, and so on. If none of the conditions are True, the code inside the else block is executed.


Example:

x = 10

if x > 15:

print("x is greater than 15")

elif x > 5:

print("x is greater than 5 but less than or equal to 15")

else:

print("x is 5 or less")

output:  x is greater than 5 but less than or equal to 15


4. Nested Conditional Statements

You can also have conditional statements inside other conditional statements. This is called nesting. It's useful when you have to check for multiple conditions in a more structured manner.

Example:

x = 10

y = 5


if x > 5:

if y > 2:

print("x is greater than 5 and y is greater than 2")

else:

print("x is greater than 5 but y is not greater than 2")

else:

print("x is 5 or less")


Output: x is greater than 5 and y is greater than 2


**Question 7. Describe the different types of loops in python and their use case with example.**

Answer:


Loops are a fundamental programming concept that allow you to execute a block of code repeatedly as long as a condition is True. In Python, there are primarily two types of loops:
* For loop
* While loop

Additionally, Python provides break, continue, and else clauses in loops to further control the flow of execution.

1. for Loop:

The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code once for each item in the sequence. It is commonly used when the number of iterations is known beforehand.

Example:

for i in range(5):   # Iterates over the range 0 to 4

print(i)

2. while Loop:

The while loop in Python repeatedly executes a block of code as long as a specified condition evaluates to True. It's commonly used when the number of iterations is not known in advance, and the loop continues until a certain condition is met.

Example:

count = 0

while count < 5:

print(count)

count += 1  # Increment count to avoid infinite loop

output:

0

3. break Statement

The break statement is used to exit a loop prematurely, regardless of the loop's condition. It's often used when a certain condition is met, and you want to exit the loop before it finishes all iterations.

Example: Using break in a for loop

for i in range(10):
if i == 5:
break  # Exit the loop when i equals 5
print(i)

4. continue Statement

The continue statement is used to skip the rest of the code inside the current iteration of a loop and move to the next iteration. It doesn't exit the loop but simply skips the current iteration and proceeds with the next one.

Example 1: Using continue in a for loop

for i in range(5):

if i == 2:

continue  # Skip the iteration when i equals 2

print(i)

5. else Clause in Loops

The else clause in loops is executed when the loop terminates normally (i.e., when the loop condition becomes False and not by a break statement). It allows you to specify code that should run after the loop finishes executing, provided the loop wasn't terminated by break.

Example 1: Using else with a for loop
