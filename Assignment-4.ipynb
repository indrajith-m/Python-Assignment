**Question No:1** What are the five key concepts of Object-Oriented Programming (OOP)?

**Answer:**

The five key concepts of Object-Oriented Programming (OOP) are:

1. Encapsulation:

Encapsulation refers to the bundling of data (attributes) and the methods (functions) that operate on the data into a single unit known as a class. This concept also involves controlling access to the data by using access modifiers (like private, protected, and public). Encapsulation helps to hide the internal implementation details and only exposes what is necessary.

2. Abstraction:

Abstraction involves simplifying complex systems by providing a high-level view while hiding the underlying implementation details. In OOP, this is achieved through abstract classes or interfaces that define the structure without specifying how the functionality is implemented. This allows for focusing on "what" an object does, rather than "how" it does it.

3. Inheritance:

Inheritance is a mechanism where one class (called the child or subclass) inherits the properties and behaviors (methods) of another class (called the parent or superclass). This allows for code reuse and the creation of a hierarchy of classes, where subclasses can override or extend the functionality of the parent class.

4. Polymorphism:

Polymorphism means "many shapes" and allows objects of different classes to be treated as objects of a common superclass. The most common use of polymorphism is through method overriding (runtime polymorphism) and method overloading (compile-time polymorphism). This allows for the same method name to behave differently depending on the object calling it.

5. Composition:

Composition involves creating complex objects by combining simpler objects. Unlike inheritance, which represents an "is-a" relationship, composition represents a "has-a" relationship. For example, a "Car" class may have a "Wheel" object, meaning a car is composed of wheels, but a wheel is not a type of car. This promotes flexibility and reusability, as objects can be composed dynamically.

These concepts work together to facilitate the creation of modular, reusable, and maintainable code in object-oriented systems.

**Question No3:** Explain the difference between instance methods and class methods. Provide an example of each.

**Answer:**

In Object-Oriented Programming (OOP), both instance methods and class methods are functions that belong to a class, but they differ in how they are called and how they interact with the class and its instances.

1. Instance Methods:

Definition: An instance method operates on an instance (an object) of the
class. It takes the instance itself as its first argument, typically named self. This allows the method to access and modify the attributes and other methods of that particular object.

Usage: Instance methods are used when you need to operate on or access the instance's data.

Example of an Instance Method:
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    # Instance method
    def display_info(self):
        print(f"Car Information: {self.year} {self.make} {self.model}")

my_car = Car("Mahindra", "XUV", 2024)
my_car.display_info()  # Calling the instance method

display_info is an instance method, as it works with the specific instance (my_car) and accesses its attributes (like make, model, and year) via self.
It is called on an object (my_car.display_info()), not directly on the class.


2. Class Methods:

Definition: A class method operates on the class itself rather than an instance of the class. It takes the class itself as the first argument, typically named cls. Class methods are often used to operate on class-level data or to create factory methods (methods that create instances of the class).

Usage: Class methods are used when you need to work with the class itself, such as accessing or modifying class-level variables or creating instances from the class.

Example of a Class Method:
class Car:
    car_count = 0  # Class attribute
    
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        Car.car_count += 1  # Incrementing car count when a new instance is created
    
    # Class method
    @classmethod
    def display_car_count(cls):
        print(f"Total number of cars: {cls.car_count}")

# Creating instances of the class
car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Honda", "Civic", 2021)

# Calling the class method
Car.display_car_count()  # Calling the class method on the class


display_car_count is a class method, as it works with the class (cls) and can access class-level data, like car_count, which tracks the total number of cars created.
It is called on the class itself (Car.display_car_count()), not on an object instance.
The class method can access and modify class attributes, whereas instance methods work with individual instance attributes.

**Answer:**


"Method Overloading":

Default Parameters: You can set default values for parameters to make a method behave differently based on the number of arguments passed.
Variable-Length Arguments: You can use *args (for a variable number of positional arguments) and **kwargs (for a variable number of keyword arguments) to handle different input combinations.
Conditional Logic: You can implement logic inside the method to handle different cases based on the type or number of arguments.

Example of "Method Overloading" in Python:

* Using Default Parameters:

class Calculator:
    # Method with default parameters
    def add(self, a, b=0, c=0):
        return a + b + c

Creating an instance of the class

calc = Calculator()

Different calls to the same method with varying arguments

print(calc.add(5))        # Only 'a' is provided (output: 5)

print(calc.add(5, 3))     # 'a' and 'b' are provided (output: 8)

print(calc.add(5, 3, 2))  # 'a', 'b', and 'c' are provided (output: 10)

Explanation: The add method in this example can be called with one, two, or three arguments due to the default values assigned to b and c. If fewer arguments are passed, the default values are used, which allows a form of overloading.

* Using *args (Variable-Length Arguments):
class Printer:
    # Method with variable-length arguments
    def print_message(self, *messages):
        for message in messages:
            print(message)

Creating an instance of the class

printer = Printer()

Different calls to the same method with varying numbers of arguments

printer.print_message("Hello!")               # One argument

Different calls to the same method with varying keyword arguments

person.greet(name="Alice")  # Output: Hello, Alice!

**Question no5:** What are the three types of access modifiers in Python? How are they denoted?

**Answer:**

In Python, access modifiers control the visibility and accessibility of class attributes and methods. While Python doesn't have strict access control mechanisms like some other languages (e.g., Java or C++), it uses a convention-based approach to define different levels of access. There are three types of access modifiers in Python:

1. Public

Description: Public members are accessible from anywhere, both inside and outside the class. These are the default type of members in Python.
Denoted: Public members do not have any leading underscores.

Example:

class MyClass:
    def __init__(self, value):
        self.public_value = value  # Public attribute

Creating an instance of MyClass

obj = MyClass(10)

Accessing the public attribute

print(obj.public_value)  # Output: 10


2. Protected

Description: Protected members are meant to be accessible only within the class and its subclasses. In Python, this is a convention rather than a strict rule, and it doesn't enforce access control. The use of protected members signals that the attribute or method should not be accessed directly from outside the class.
Denoted: A single underscore (_) before the attribute or method name.


Example:

class MyClass:
    def __init__(self, value):
        self._protected_value = value  # Protected attribute

class SubClass(MyClass):
    def display(self):
        print(self._protected_value)  # Accessing protected attribute within subclass

Creating an instance of MyClass

obj = MyClass(10)

Accessing the protected attribute outside the class (not recommended)

print(obj._protected_value)  # Output: 10 (but it's discouraged)

3. Private

Description: Private members are intended to be accessible only within the class itself. They are not meant to be accessed from outside the class or even from subclasses. This is more of a name-mangling convention in Python, where the name of the attribute or method is modified internally to prevent accidental access.
Denoted: A double underscore (__) before the attribute or method name.

Example:

class MyClass:
    def __init__(self, value):
        self.__private_value = value  # Private attribute

    def __private_method(self):
        print("This is a private method.")

    def display(self):
        print(self.__private_value)  # Accessing private attribute within the class

Creating an instance of MyClass

obj = MyClass(10)

Accessing private attribute directly (raises AttributeError)

print(obj.__private_value)  # AttributeError: 'MyClass' object has no attribute '__private_value'

Accessing the private method directly (raises AttributeError)

obj.__private_method()  # AttributeError: 'MyClass' object has no attribute '__private_method'

Accessing private members using name mangling (not recommended)

print(obj._MyClass__private_value)  # Output: 10


__private_value and __private_method are private, and trying to access them directly from outside the class results in an AttributeError.
However, Python "mangles" the name of private attributes by prefixing the class name (_MyClass__private_value), making it possible (though not recommended) to access them this way.

**Question No6:** Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

**Answer:**

In Python, inheritance allows a class (child or subclass) to inherit properties and behaviors (methods) from another class (parent or superclass). Python supports different types of inheritance, and these types can be used to create complex class hierarchies.

1. Single Inheritance

Definition: A subclass inherits from a single parent class. The subclass can access all public and protected attributes and methods of the parent class.

Example: A class Dog inherits from a class Animal.


Example of Single Inheritance:
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")


Creating an instance of Dog

dog = Dog()

dog.speak()  # Inherited from Animal

dog.bark()   # Defined in Dog

2. Multiple Inheritance

Definition: A subclass inherits from more than one parent class. The subclass can access attributes and methods from all its parent classes. This allows the subclass to combine functionalities from multiple classes.
Example: A class Student inherits from both Person and Employee classes.

Example of Multiple Inheritance:

class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        print(f"Hello, my name is {self.name}")

class Employee:
    def __init__(self, job_title):
        self.job_title = job_title
    
    def work(self):
        print(f"Working as a {self.job_title}")

class Student(Person, Employee):
    def __init__(self, name, job_title, school):
        Person.__init__(self, name)  # Initializing parent class Person
        Employee.__init__(self, job_title)  # Initializing parent class Employee
        self.school = school
    
    def study(self):
        print(f"Studying at {self.school}")

Creating an instance of Student

student = Student("Alice", "Software Developer", "XYZ University")

student.greet()    # From Person

student.work()     # From Employee

student.study()    # From Student

Explanation:

The Student class inherits from both Person and Employee classes.
It calls the constructors (__init__) of both parent classes to initialize the name and job_title.
Student can now access methods from both Person (greet) and Employee (work), and it also defines its own method study.

3. Multilevel Inheritance

Definition: In multilevel inheritance, a subclass inherits from a parent class, and another class inherits from the subclass, creating a multi-tier inheritance chain.
Example: A class Grandchild inherits from Child, and Child inherits from Parent.

Example of Multilevel Inheritance:

class Parent:
    def speak(self):
        print("Parent speaks")

class Child(Parent):
    def talk(self):
        print("Child talks")

class Grandchild(Child):
    def chat(self):
        print("Grandchild chats")

Creating an instance of Grandchild

grandchild = Grandchild()

grandchild.speak()  # Inherited from Parent

grandchild.talk()   # Inherited from Child

grandchild.chat()   # Defined in Grandchild

Explanation:
Grandchild inherits from Child, and Child inherits from Parent.
Grandchild can access methods from both Parent and Child in addition to its own method.

4. Hierarchical Inheritance

Definition: In hierarchical inheritance, multiple subclasses inherit from a single parent class. All subclasses share the parent class's attributes and methods.
Example: Both Dog and Cat classes inherit from the Animal class.

Example of Hierarchical Inheritance:

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

Creating instances of Dog and Cat

dog = Dog()

cat = Cat()

dog.speak()  # Inherited from Animal

dog.bark()   # Defined in Dog

cat.speak()  # Inherited from Animal

cat.meow()   # Defined in Cat

Explanation:
Both Dog and Cat inherit from Animal, so they can access the speak method from Animal.
Each subclass can also define its own specific methods (bark for Dog and meow for Cat).

5. Hybrid Inheritance

Definition: Hybrid inheritance is a combination of two or more types of inheritance. It can involve multiple inheritance, multilevel inheritance, or hierarchical inheritance within a single class hierarchy.
Example: A class C inherits from both a subclass and a parent class, combining multiple and multilevel inheritance.

Example of Hybrid Inheritance:

class Parent:
    def speak(self):
        print("Parent speaks")

class Child(Parent):
    def talk(self):
        print("Child talks")

class AnotherClass:
    def greet(self):
        print("Another class greeting")

class HybridChild(Child, AnotherClass):
    def hello(self):
        print("HybridChild says hello")

Creating an instance of HybridChild

hybrid = HybridChild()

hybrid.speak()   # Inherited from Parent

hybrid.talk()    # Inherited from Child

hybrid.greet()   # Inherited from AnotherClass

hybrid.hello()   # Defined in HybridChild

Explanation:
HybridChild class inherits from both Child (which is derived from Parent) and AnotherClass.
It can access methods from Parent, Child, and AnotherClass.

Multiple Inheritance Example:

In the Multiple Inheritance example provided above, the Student class inherits from both Person and Employee, allowing it to access both the greet and work methods, while also defining its own study method. This illustrates how multiple inheritance enables a class to combine functionalities from different classes.

**Question No7:** What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

**Answer:**

Method Resolution Order (MRO) in Python

In Python, the Method Resolution Order (MRO) is the order in which the methods and attributes of a class are inherited in the case of multiple inheritance. It determines the sequence in which classes are looked up when calling a method or accessing an attribute. The MRO is important in a multiple inheritance scenario, where a class inherits from more than one parent class.

The MRO ensures that:

* Each class is only visited once.
* The order in which base classes are searched follows a well-defined algorithm.

C3 Linearization (C3 Superclass Linearization):

Python uses a specific algorithm called C3 Linearization to determine the method resolution order in the case of multiple inheritance. This algorithm ensures that classes are visited in a consistent order, taking into account the inheritance hierarchy and the order in which base classes are listed in the subclass definition.

The C3 linearization process follows these rules:

* Depth-first search: Start from the subclass and look at the classes listed in its inheritance.
* Parents are searched before their children: Parents should appear before their descendants in the resolution order.
* Maintains order of inheritance: The order of base classes in the subclass declaration is respected.

How MRO Works

For example, in the case of multiple inheritance, the MRO ensures that the base classes are searched in the correct order, taking into account the hierarchy of classes.

Example of MRO in Multiple Inheritance

class A:
    def method(self):
        print("Method in class A")

class B(A):
    def method(self):
        print("Method in class B")

class C(A):
    def method(self):
        print("Method in class C")

class D(B, C):
    pass

Creating an instance of class D

d = D()

d.method()

In this example, class D inherits from both B and C, which in turn inherit from A. When d.method() is called, Python will search for the method in the order defined by the MRO.
The MRO will determine that the method should be searched in the order: D -> B -> C -> A.

The output will be: Method in class B

How to Retrieve the MRO Programmatically:

You can retrieve the MRO of a class using the mro() method or the __mro__ attribute.

1. Using mro() Method

The mro() method is available on a class and returns a list of classes in the method resolution order.

print(D.mro())

This will return a list of classes in the MRO for class D:

[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]

2. Using __mro__ Attribute

Alternatively, you can access the __mro__ attribute, which is a tuple of classes in the method resolution order.

print(D.__mro__)

This will give the same result: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)

Explanation of the MRO Output

The output shows the classes in the order Python will search for the method starting from D and going up the inheritance chain.
The last class in the MRO is always the object class, which is the base class of all classes in Python.

Why MRO is Important

Consistent Inheritance: It helps ensure a consistent, predictable order when Python searches for a method or attribute in the class hierarchy.
Multiple Inheritance: MRO plays a crucial role in determining how classes interact with each other in multiple inheritance situations. Without it, the order of method calls and attribute lookups could become ambiguous and error-prone.

Summary

The Method Resolution Order (MRO) in Python determines the order in which classes are searched for methods and attributes, particularly in the case of multiple inheritance.
Python uses the C3 Linearization algorithm to establish a consistent and predictable MRO.
You can retrieve the MRO using the mro() method or the __mro__ attribute of a class.


**Question No8:** Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses
`Circle` and `Rectangle` that implement the `area()` method.


**Answer:**


```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

circle = Circle(10)
rectangle = Rectangle(5, 10)

print(f"Area of Circle: {circle.area()}")
print(f"Area of Rectangle: {rectangle.area()}")
```

**Question No9:** Demonstrate polymorphism by creating a function that can work with different shape objects to calculate
and print their areas.

**Answer:**

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


def print_area(shape: Shape):
    print(f"The area of the {shape.__class__.__name__} is: {shape.area()}")


circle = Circle(10)
rectangle = Rectangle(5, 10)
triangle = Triangle(15, 20)

# Use the print_area function with polymorphism
print_area(circle)
print_area(rectangle)
print_area(triangle)
```

2. Subclasses (Circle, Rectangle, Triangle):

Each shape class inherits from Shape and implements the area() method with its specific formula.

Circle calculates area using 2, Rectangle uses width  height, and Triangle uses 0.5  base  height.


3. Polymorphic Function (print_area):

The function print_area() takes any object that is an instance of Shape and calls the area() method on it, displaying the area of that specific shape.

This demonstrates polymorphism because the same function (print_area) works with different shape objects (Circle, Rectangle, Triangle), even though each shape has a different implementation for calculating its area.

**Question No:10 ** Implement encapsulation in a `BankAccount` class with private attributes for `balance` and
`account_number`. Include methods for deposit, withdrawal, and balance inquiry.

**Answer:**


```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0.0):
        self.__account_number = account_number
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance is ${self.__balance}.")
        else:
            print("Deposit amount must be greater than zero.")


    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance is ${self.__balance}.")
        elif amount > self.__balance:
            print("Insufficient funds for this withdrawal.")
        else:
            print("Withdrawal amount must be greater than zero.")

    def get_balance(self):
        return self.__balance


    def get_account_number(self):
        return self.__account_number

account = BankAccount("123456789", 1000.0)

account.deposit(10000)
account.withdraw(2500)
print(f"Account balance: ${account.get_balance()}")
```

1. Private Attributes:

The __balance and __account_number attributes are private, indicated by the double underscore (__) at the beginning of the attribute names. This prevents direct access to these attributes from outside the class.

2. Methods:

Deposit(amount) allows a deposit to the account if the amount is positive. It updates the balance and prints the new balance.
withdraw(amount) allows a withdrawal if the amount is positive and does not exceed the balance. It checks for sufficient funds before updating the balance.

Get_balance() returns the current balance of the account.
get_account_number() provides access to the account number (this method could be omitted or used in a controlled way if needed).


3. Encapsulation:

The private attributes cannot be directly accessed outside the class, ensuring that only methods within the class can modify or retrieve them.
The public methods (deposit, withdraw, get_balance) provide controlled access to the balance, ensuring that deposits and withdrawals are properly validated.

1. __str__:

The __str__ method allows you to control the string representation of your object. This is useful for debugging and when you want a more user-friendly or meaningful string output when printing the object or converting it to a string.


2. __add__:

The __add__ method allows you to define the behavior of the + operator between objects of your class. In this case, it allows us to add two Point objects by adding their respective coordinates, which is useful when you want to combine or merge data encapsulated in objects.

We override the __add__ method to define how two Point objects are added together. When we use the + operator between two Point objects, Python calls the __add__ method. In our implementation, the x and y coordinates of the two points are added together, and a new Point object with the resulting coordinates is returned.

**Question No12:**  Create a decorator that measures and prints the execution time of a function.

**Answer:**

To create a decorator that measures and prints the execution time of a function, we can use Python's time module to capture the start and end times. The decorator will wrap the original function, record the time before and after its execution, and then print the duration.

What the Decorator Allows You to Do:
1. Measure Execution Time: This decorator wraps any function and prints the time it takes to execute. It's useful for profiling and optimizing code, especially in performance-sensitive applications.

2. Reusable and Clean: The decorator can be applied to any function without modifying the function's original code. Just add @measure_time above the function definition.

```python
import time

def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution time of {func.__name__}: {execution_time:.6f} seconds")
        return result
    return wrapper

@measure_time
def slow_function():
    time.sleep(2)

@measure_time
def fast_function():
    time.sleep(0.5)

slow_function()
fast_function()
```

**Question No13:**  Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

**Answer:**

        A
       / \
      B   C
       \ /
        D

In this structure:

Class A is the root or base class.
Classes B and C both inherit from class A.
Class D inherits from both B and C.
The Problem:
The problem arises when class D inherits from both B and C, and both B and C inherit from A. In this situation, if there is a method in A that is overridden in both B and C, and D calls that method, it creates ambiguity about which method should be called  the one from B, C, or A.

For example:

If class A has a method do_something(), and both B and C override this method, when class D calls do_something(), it's unclear whether to use the method from B, C, or A.
This situation leads to confusion and unexpected behavior, which is the Diamond Problem.

Example of the Diamond Problem:

class A:
    def do_something(self):
        print("A's method")

class B(A):
    def do_something(self):
        print("B's method")

class C(A):
    def do_something(self):
        print("C's method")

class D(B, C):
    pass

Creating an instance of D

d = D()

d.do_something()  # Which method will be called? B's or C's?

In this example, class D inherits from both B and C, which in turn inherit from A. The question is: which do_something() method will be called when we call d.do_something()?

How Python Resolves the Diamond Problem

Python resolves the Diamond Problem using a method resolution order (MRO) algorithm called C3 Linearization (or C3 superclass linearization). This algorithm ensures a consistent order for method resolution in the case of multiple inheritance.

Key Points of C3 Linearization:

* The MRO is a linear order in which base classes are considered when searching for a method.
* The MRO for each class is calculated based on the inheritance hierarchy, and it ensures that classes are checked in a consistent and predictable order.
* MRO is determined once for each class, and it follows the inheritance structure defined by the programmer.
* Python uses the mro() method (or __mro__ attribute) to view the order in which methods will be resolved.

Example with MRO in Python:

class A:
    def do_something(self):
        print("A's method")

class B(A):
    def do_something(self):
        print("B's method")

class C(A):
    def do_something(self):
        print("C's method")

class D(B, C):
    pass

Create an instance of D

d = D()

d.do_something()

Display the MRO

print(D.__mro__)

Output:
B's method
(<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)


1. Method Resolution:

* When we call d.do_something(), Python will check the MRO for D.
* The MRO of D is: [D, B, C, A, object].
* Python first checks D, then B, and since B has the do_something() method, it calls B's method.

2. MRO of D:

* he MRO of D is computed as follows:
* Python starts with D itself.
* Then it looks at the parents of D  first B, then C.
* Since both B and C inherit from A, Python checks A next.
* Finally, it ends with object, which is the base class of all Python classes.

In the case of the Diamond Problem, C3 Linearization ensures that B is considered before C because B is listed first in the class definition of D. This is the consistent method resolution order.

3. The MRO can be explicitly checked using __mro__, which shows the exact order in which methods will be looked up for a given class.

Key Takeaways:

Diamond Problem arises in multiple inheritance when a class inherits from two classes that both inherit from a common base class.
C3 Linearization resolves the Diamond Problem in Python by determining a consistent method resolution order (MRO).
Python ensures that the method resolution follows a predictable order and avoids ambiguity by checking the classes in a linear order based on the inheritance hierarchy.

**Question No14:**  Write a class method that keeps track of the number of instances created from a class.

**Answer:**

To keep track of the number of instances created from a class, you can use a class variable that is shared among all instances of the class. A class method can then be used to access and update this variable each time an instance is created.


Class Variable: A variable that is shared among all instances of a class.
Class Method: A method that is bound to the class and not the instance. It can be used to modify class variables.

Example Code:

class MyClass:
    # Class variable to track the number of instances
    instance_count = 0

    def __init__(self):
        # Increment the instance count whenever a new instance is created
        MyClass.instance_count += 1

    @classmethod
    def get_instance_count(cls):
        # Class method to return the current count of instances
        return cls.instance_count

Example usage:

Creating instances of MyClass

obj1 = MyClass()

obj2 = MyClass()

obj3 = MyClass()

Accessing the class method to get the instance count

print(MyClass.get_instance_count())

Output: 3

Explanation:
1. instance_count (Class Variable):

* This variable is declared at the class level and is shared by all instances of the class. It starts with a value of 0 and is incremented every time a new instance of MyClass is created.

2. __init__ (Constructor):

* In the constructor (__init__), each time a new instance is created, the class variable instance_count is incremented by 1.

3. get_instance_count (Class Method):

* This is a class method, which is defined using the @classmethod decorator. It is used to access the class variable instance_count and return the number of instances created so far. The cls parameter is used to refer to the class itself (as opposed to an instance).

4. Creating Instances:

* When instances obj1, obj2, and obj3 are created, the instance_count is incremented each time.

5. Accessing Instance Count:

* The method get_instance_count() is called using the class name (MyClass.get_instance_count()) to get the number of instances created.

**Question No15:**   Implement a static method in a class that checks if a given year is a leap year.

**Answer:**

To implement a static method in a class that checks if a given year is a leap year, we need to follow the standard rules for leap years:

Leap Year Rules:

* A year is a leap year if it is divisible by 4.
* However, if the year is divisible by 100, it is not a leap year, unless:
* The year is also divisible by 400, in which case it is a leap year.

Approach:
* The static method will not depend on any instance-specific data (i.e., it does not require access to self or cls), so we will use the @staticmethod decorator.
* The static method will take the year as an argument and return True if the year is a leap year, or False otherwise.

Code Implementation:

class YearUtils:
    @staticmethod
    def is_leap_year(year):
        # Check if the year is a leap year using the leap year rules
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        else:
            return False

Example usage:

print(YearUtils.is_leap_year(2020))  # Output: True (2020 is a leap year)

print(YearUtils.is_leap_year(2024))  # Output: True (2024 is a leap year)

print(YearUtils.is_leap_year(1900))  # Output: False (1900 is not a leap year)

print(YearUtils.is_leap_year(2000))  # Output: True (2000 is a leap year)


Explanation:
1. Static Method:

* The method is_leap_year is defined with the @staticmethod decorator, meaning it does not require an instance (self) or class (cls) to be called.
* It simply takes year as an argument and performs the leap year check.

2. Leap Year Logic:

* The method checks:
* If the year is divisible by 4 and not divisible by 100, it returns True.
* Alternatively, if the year is divisible by 400, it also returns True.
* If neither condition is met, it returns False.

3. Example Usage:

* You can call the static method directly on the class (YearUtils.is_leap_year(2020)), without creating an instance of YearUtils.
* It will return True for leap years and False for non-leap years.

Key Points:
* Static Method: The @staticmethod decorator indicates that the method does not require access to instance or class-level data.
* Leap Year Check: The method checks the rules of leap years and returns the appropriate result.
