**Question No:1**  Discuss String Slicing and provide exapmle.

**Answer:**

String slicing is a powerful feature in Python that allows you to extract a portion of a string using a specified range of indices. In Python, strings are ordered sequences, and string slicing uses the : operator to access a subsequence from a given string. You can specify a start, stop, and step to slice the string.

**start:** The index where the slice starts (inclusive).

**stop:** The index where the slice ends (exclusive).

**step:** The step between each index, which can be used to skip characters (optional).



**Default Values:**

If start is omitted, it defaults to 0 (the beginning of the string).

If stop is omitted, it defaults to the length of the string (i.e., the slice will go until the end of the string).

If step is omitted, it defaults to 1 (i.e., no skipping).


Example:
1.Basic Slicing

text = "Hello, World!"

Extracts characters from index 0 to 4 (inclusive)
slice1 = text[0:5]  # "Hello"

Extracts characters from index 7 to 11 (exclusive of 12)
slice2 = text[7:12]  # "World"

print(slice1)  # Output: Hello

print(slice2)  # Output: World

**Question No:2** Explain the key features of list in python?

**Answer:**

List is one of the most versatile and widely used data structures. It is an ordered collection of elements, which can hold items of different types (including other lists). Lists are mutable, meaning that their elements can be changed after creation.


Summary of List Key Features:

* Ordered: Elements are stored in a sequence.

* Mutable: You can change the elements after creation.

* Heterogeneous: Can store elements of different data types.

* Indexed: You can access elements using indices.

* Slicing: You can extract sublists using slicing.

* Dynamic: Lists can grow or shrink in size.

* Supports Nesting: Lists can contain other lists.

* Iterable: You can loop through lists.
* Various Methods: Python provides built-in methods for list manipulation.

* Duplicates Allowed: Lists can have repeated elements.




1. Ordered

* Lists maintain the order of elements in which they were added. The first element has index 0, the second element has index 1, and so on.

* The order in which elements are inserted is preserved when accessed or iterated over.

Example:

my_list = [10, 20, 30]

print(my_list[0])

Output: 10

2. Mutable
* Lists are mutable, meaning you can modify, add, or remove elements after the list is created.

Example:
my_list = [10, 20, 30]

my_list[1] = 99  # Modifying the second element

print(my_list)  # Output: [10, 99, 30]

my_list.append(40)  # Adding an element at the end

print(my_list)  # Output: [10, 99, 30, 40]

my_list.remove(99)  # Removing an element

print(my_list)  # Output: [10, 30, 40]

3. Heterogeneous
* Lists can contain elements of different types: integers, strings, floats, or even other lists (nested lists).

Example:

mixed_list = [1, "hello", 3.14, [2, 4, 6]]

print(mixed_list)  # Output: [1, 'hello', 3.14, [2, 4, 6]]

4. Indexed
* List elements are accessed via zero-based indexing, meaning you can use an integer index to access each item. Negative indexing allows accessing elements from the end of the list.

Example:

my_list = ['apple', 'banana', 'cherry']

print(my_list[1])  # Output: banana

print(my_list[-1]) # Output: cherry (last element)


5. Slicing
* Lists support slicing, which allows you to extract sublists (portions of the list) using the start:stop:step syntax.

Example:

my_list = [10, 20, 30, 40, 50]

slice1 = my_list[1:4]  # Extract elements from index 1 to 3

print(slice1)  # Output: [20, 30, 40]


slice2 = my_list[::2]  # Every second element

print(slice2)  # Output: [10, 30, 50]

6. Dynamic Size
* Lists are dynamic, meaning their size can grow or shrink as elements are added or removed, without needing to declare the size beforehand.

Example:

my_list = [1, 2, 3]

my_list.append(4)  # Adding an element

print(my_list)  # Output: [1, 2, 3, 4]

my_list.pop()  # Removing the last element

print(my_list)  # Output: [1, 2, 3]

7. Supports Nesting
* Lists can contain other lists, allowing for multi-dimensional lists (like matrices or tables).

Example:

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(matrix[0])  # Output: [1, 2, 3] (first row)

print(matrix[0][1])  # Output: 2 (second element of the first row)

8. Iterable
* Lists are iterable objects, meaning you can loop through the elements using loops such as for or while.

Example:

my_list = [1, 2, 3]

for item in my_list:
print(item)
Output:

1

2

3

9. Support for Various Methods
* Lists come with a variety of built-in methods that make them highly flexible for manipulation. Some common methods include:

append(item): Adds an item to the end of the list.

extend(iterable): Adds elements from an iterable (like another list) to the end of the list.

insert(index, item): Inserts an item at a specific index.

remove(item): Removes the first occurrence of the item from the list.

pop([index]): Removes and returns the element at the specified index. If no index is specified, it removes and returns the last element.

index(item): Returns the index of the first occurrence of the item.

count(item): Returns the number of occurrences of an item.

sort(): Sorts the list in ascending order (can be customized with reverse=True).

reverse(): Reverses the order of elements in the list.

clear(): Removes all elements from the list.

Example:

my_list = [3, 1, 4, 1, 5, 9]

my_list.append(2)  # Add 2 to the end

my_list.sort()     # Sort the list

print(my_list)     # Output: [1, 1, 2, 3, 4, 5, 9]


10. Duplicates Allowed
* Lists can have duplicate elements, meaning the same value can appear multiple times in a list.

Example:

my_list = [1, 2, 2, 3, 3, 3]

print(my_list)  # Output: [1, 2, 2, 3, 3, 3]

**Question No:4**  Comapre and contrast tuples and list with examples

**Answer:**

Both tuples and lists are widely used collection types in Python, and while they share some similarities (e.g., both can hold ordered collections of items, and both allow indexing), they differ significantly in key aspects such as mutability, syntax, and use cases. Here's a detailed comparison of tuples and lists, highlighting their differences and similarities.

1. Syntax

* List: Defined using square brackets [].

* Tuple: Defined using parentheses ().

Example:

List

my_list = [1, 2, 3, 4]

Tuple

my_tuple = (1, 2, 3, 4)

2. Mutability
* List: Lists are mutable, meaning their elements can be changed, added, or removed after the list is created.
* Tuple: Tuples are immutable, meaning once a tuple is created, its elements cannot be changed, added, or removed.

Example (Mutability):

List is mutable

my_list = [10, 20, 30]

my_list[1] = 99  # Modifying the second element
print(my_list)  # Output: [10, 99, 30]

Tuple is immutable

my_tuple = (10, 20, 30)

my_tuple[1] = 99  # This will raise a TypeError: 'tuple' object does not support item assignment


3. Performance
* List: Lists are generally slower than tuples when it comes to iteration and access because of their mutability.
* Tuple: Tuples are faster than lists in terms of iteration and access due to their immutability and fixed size.

4. Use Cases
* List: Lists are typically used when you need a mutable sequence, meaning the data might change during the program's execution. Lists are often used for storing homogeneous data that needs to be modified, such as in loops or data processing tasks.

* Tuple: Tuples are generally used for immutable data, where you want to ensure that the values do not change. Tuples are often used for storing heterogeneous data (e.g., coordinates, fixed collections of values, function return values) and are frequently used as keys in dictionaries (because they are hashable).

5. Methods Available
* List: Lists have more built-in methods because they are mutable. Common methods include:

append(): Add an element to the end.

remove(): Remove an element by value.

pop(): Remove and return an element at a specific index.

extend(): Add elements from another iterable.

insert(): Insert an element at a specific position.

sort(): Sort the list in place.

* Tuple: Tuples have fewer methods because they are immutable. Common methods include:

count(): Count occurrences of a value in the tuple.

index(): Find the index of the first occurrence of a value.

6. Memory Efficiency

* List: Lists are less memory-efficient because they are designed to be more flexible (mutable), meaning they need extra space to accommodate changes like adding/removing elements.
* Tuple: Tuples are more memory-efficient than lists because they are fixed-size, and Python can allocate memory more efficiently for them

7. Immutability and Hashing
* List: Since lists are mutable, they are not hashable, meaning they cannot be used as dictionary keys or elements of a set.
* Tuple: Since tuples are immutable, they are hashable (provided all their elements are hashable). This makes tuples suitable for use as keys in dictionaries or elements in sets.

8. Packing and Unpacking
* List: Lists can be used to pack and unpack data, but they are generally not used as a method for grouping data in a fixed, unchanging manner.
* Tuple: Tuples are frequently used for packing and unpacking data. Python allows you to return multiple values from a function as a tuple and then unpack it directly.

9. Slicing
* Both lists and tuples support slicing (i.e., extracting a part of the sequence), and they behave similarly in this regard.

Key Features of Sets in Python:

A set in Python is an unordered collection of unique elements. Unlike lists and tuples, sets are not indexed and do not allow duplicates. Here are the key features of sets:

1. Unordered Collection:

* Sets are unordered, meaning the elements are not stored in a particular order. When you print a set, the order of the elements may not be the same as when they were added.

Example:
python
Copy code
my_set = {3, 1, 2}

print(my_set)  # Output: {1, 2, 3} (order is not guaranteed)

2. Unique Elements:


* Sets do not allow duplicate elements. If you try to add a duplicate element, it will simply be ignored.

Example:
my_set = {1, 2, 3, 3, 2, 1}

3. Mutable:
* Sets are mutable, meaning you can add or remove elements after the set is created. However, the elements inside the set must be immutable (e.g., numbers, strings, or tuples).

Example:

my_set = {1, 2, 3}

my_set.add(4)

print(my_set)  # Output: {1, 2, 3, 4}

Removing an element

my_set.remove(2)

print(my_set)  # Output: {1, 3, 4}

print(my_set)  # Output: {1, 2, 3} (duplicates are removed)

* Since sets are unordered, you cannot use indexing, slicing, or access by position as you would with lists or tuples.

Example:
my_set = {1, 2, 3}

Symmetric Difference (^ or .symmetric_difference()): Returns elements that are in one set but not both.


Example:

8. Mathematical Set Operations:
* Sets in Python can perform mathematical set operations like subset, superset, and disjoint.


Subset (<= or .issubset()): Checks if all elements of one set are in another set.

Superset (>= or .issuperset()): Checks if all elements of another set are in the first set.

Disjoint (isdisjoint()): Checks if two sets have no elements in common.

**Question No:8**  Discuss the importance of dictionary keys being immutable and provide example.

**Answer:**

Importance of Dictionary Keys Being Immutable:

* In Python, dictionary keys must be immutable objects, meaning that the key itself cannot be changed after it is created. This is crucial for several reasons, including ensuring data integrity and enabling efficient lookups.

Why Dictionary Keys Must Be Immutable:
* Hashing and Efficiency: Dictionaries in Python are implemented using hash tables. In order for a dictionary to quickly look up a value associated with a key, it needs to compute the hash value of the key. If the key could be modified after insertion, its hash value could change, which would lead to issues when trying to find the key later.

* If a key is mutable (e.g., a list), and we modify it after its been used in a dictionary, the dictionary would have no way of finding it because the hash value would no longer match the original hash stored in the dictionary.

* Data Integrity: Making keys immutable ensures that the data structure remains consistent and reliable. Allowing mutable objects (like lists) to be used as keys would potentially introduce unexpected behavior, as the key could change while it is being used in the dictionary, leading to erroneous behavior.

* Consistency of Hash Function: A keys hash value must be consistent throughout its lifetime. When a key is immutable, its hash value stays constant because the data it contains cannot change. This is a key requirement for the efficient operation of hash-based data structures like dictionaries.

Examples of Immutable and Mutable Keys:

Example 1:
Immutable Keys (Allowed)
Common immutable data types that can be used as dictionary keys include strings, tuples, and numbers.

Immutable keys (string, tuple, integer)

my_dict = {"name": "Alice",(1, 2): "Coordinates",  # Tuple as a key
100: "Value"}


print(my_dict)
Output: {'name': 'Alice', (1, 2): 'Coordinates', 100: 'Value'}

In the example above:

A string ("name") is used as a key.

A tuple (1, 2) is used as a key, which is allowed because tuples are immutable.

An integer (100) is also used as a key, and integers are immutable as well.

Example 2:
Mutable Keys (Not Allowed)

If you try to use a mutable object (like a list) as a dictionary key, Python will raise a TypeError, as lists are mutable and their hash value can change.

Mutable key (list) - This will raise an error
my_dict = { [1, 2, 3]: "This will fail"  # List is mutable and cannot be used as a key }
Output TypeError: unhashable type: 'list'

In this case, attempting to use a list as a key results in a TypeError because lists are mutable, and their hash value can change, making them unsuitable for use as dictionary keys.
