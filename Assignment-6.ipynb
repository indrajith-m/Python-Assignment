**Question No2:** How does broadcasting work in NumPy?

**Answer:**

Broadcasting in NumPy:

Broadcasting is a powerful feature in NumPy that allows you to perform operations on arrays of different shapes without explicitly reshaping them. It automatically expands the smaller array to match the shape of the larger array, making element-wise operations more efficient and simpler.

In essence, broadcasting allows NumPy to apply binary operations (like addition, multiplication, etc.) to arrays with different shapes in a way that makes sense without needing to duplicate data.

How Broadcasting Works:

Broadcasting follows a set of rules to determine how arrays are aligned for element-wise operations. If the shapes of the arrays do not match, NumPy will "stretch" the smaller array to match the larger one, without copying data, but by repeating it virtually.

Here’s the basic idea:

* Starting from the right-most dimension, compare the shapes of the two arrays element by element.
* If the sizes in that dimension are the same, or if one of them is 1, they are compatible.
* If the sizes are different and neither is 1, broadcasting will not be possible, and an error will occur.

Example of Broadcasting:


Case 1: Simple Scalar and Array Operation

Let's look at a simple case where broadcasting happens between a scalar and a NumPy array.

import numpy as np

Scalar

scalar = 3

Array

array = np.array([1, 2, 3, 4])

Broadcasting the scalar to the array (adding 3 to each element)

result = array + scalar

print(result)

Output: [4,5,6,7]


Case 2: Arrays of Different Shapes

import numpy as np

2D Array

array_2d = np.array([[1, 2], [3, 4]])

1D Array

array_1d = np.array([10, 20])

Broadcasting the 1D array to match the shape of the 2D array

result = array_2d + array_1d

print(result)

Output: [[11 22]
        [13 24]]

In this example, the 1D array [10, 20] is broadcast across the 2D array [[1, 2], [3, 4]]. The second array is "stretched" to match the shape of the first array:

10 is added to each element in the first column.
20 is added to each element in the second column.

Case 3: Arrays with Shapes That Are Not Automatically Compatible
Let’s take a case where broadcasting won’t work.

import numpy as np

Arrays with incompatible shapes

array_1 = np.array([1, 2, 3])

array_2 = np.array([[1, 2], [3, 4]])

Trying to add arrays of incompatible shapes

result = array_1 + array_2  # This will raise an error

ValueError: operands could not be broadcast together with shapes (3,) (2,2)

Here, array_1 has a shape of (3,) and array_2 has a shape of (2, 2). These shapes are not compatible for broadcasting, so NumPy raises an error.


Rules of Broadcasting:

If the arrays have a different number of dimensions, the smaller-dimensional array is padded with ones on the left side until both arrays have the same number of dimensions.

For example, an array with shape (3,) can be broadcast with an array of shape (1, 3) because NumPy will pad the smaller array to have a shape of (1, 3).

The dimensions are compared element-wise starting from the right (the last dimension). Two dimensions are compatible when:

They are the same size, or One of them is 1, in which case the smaller array will be "stretched" to match the larger array.

If the dimensions are incompatible, broadcasting will not be possible, and a ValueError is raised.

**Question No3:**  What is a Pandas DataFrame?

A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure used for data manipulation and analysis in Python. It is one of the core data structures in the Pandas library, which is built on top of NumPy. A DataFrame is similar to a table or a spreadsheet in which data is organized in rows and columns.

Key Features of a Pandas DataFrame:

1. Two-Dimensional Structure:

A DataFrame consists of rows and columns, making it similar to a table or an SQL database. Each column can have a different data type (e.g., integers, floats, strings), and each row can represent a different observation or record.

2. Heterogeneous Data:

Unlike NumPy arrays, where all elements must have the same data type, a DataFrame can hold different data types in different columns. For example, one column can have integers, another can have strings, and a third could contain floating-point numbers.

3. Indexing:

A DataFrame has both a row index and a column index. The row index labels the rows (which can be integers, strings, or other types), and the column index labels the columns. If not specified, Pandas automatically assigns default integer indices starting from 0.
Indexes allow for easy access and selection of data in the DataFrame.

4. Column Access:

You can access and manipulate columns of a DataFrame like Python dictionaries. Each column can be accessed using the column label, and individual columns are usually stored as Pandas Series (which is a one-dimensional array).

5. Data Alignment:

Pandas automatically aligns data when performing operations between DataFrames, ensuring that the data from different sources is correctly aligned by the index.

6. Handling Missing Data:

Pandas provides easy handling of missing data (NaN values). There are built-in functions to detect, remove, or fill in missing data in a DataFrame.

Why Pandas DataFrames are Useful:

Data Analysis: DataFrames are widely used in data analysis and manipulation tasks. They provide tools for cleaning, transforming, and summarizing data in an easy-to-understand format.

Integration with Other Libraries: Many Python libraries (such as Matplotlib, Seaborn, and scikit-learn) integrate seamlessly with Pandas DataFrames, making it easy to transition between data manipulation and analysis.

Efficient Handling of Large Datasets: While Pandas can handle large datasets, it also offers tools for filtering, grouping, and aggregating data, allowing you to efficiently explore and analyze large volumes of information.

Flexible Data Representation: The DataFrame structure allows for the representation of complex data in a simple table format, which makes it easier to work with and interpret.

**Question No6:**  What are the differences between NumPy arrays and Python lists?

**Answer:**

There are several important differences between NumPy arrays and Python lists, especially in terms of functionality, performance, and memory efficiency. Below is a comparison that outlines the key differences:

1. Data Type Homogeneity

Python Lists:

A Python list can store elements of different data types (integers, strings, floats, etc.).
Example: my_list = [1, 2, 'three', 4.0]

NumPy Arrays:

A NumPy array, on the other hand, requires all elements to have the same data type. This makes NumPy arrays more efficient in terms of memory and operations.
Example: np_array = np.array([1, 2, 3]) (all elements are integers)

Conclusion: NumPy arrays are homogeneous, while Python lists are heterogeneous.

2. Memory Efficiency

Python Lists:

Python lists are flexible and general-purpose containers, but they store data in a way that is not optimized for numerical computations.
As a result, Python lists consume more memory, especially when the data is large.

NumPy Arrays:

NumPy arrays are more memory-efficient. Since they store data in a contiguous block of memory and use a fixed-size data type, they are more compact and optimized for handling large datasets.

Conclusion: NumPy arrays are more memory-efficient than Python lists.

3. Performance (Speed)

Python Lists:

Python lists are slower when performing numerical operations, especially when dealing with large datasets, because they are not optimized for vectorized operations.

NumPy Arrays:

NumPy arrays are much faster because they are implemented in C and allow vectorized operations (performing operations on entire arrays at once) rather than looping through individual elements like Python lists. This significantly improves performance in numerical computations.

Conclusion: NumPy arrays are much faster for numerical operations than Python lists.

4. Mathematical Operations

Python Lists:

Python lists do not directly support mathematical operations such as addition, subtraction, multiplication, etc., on the entire list without using loops or list comprehensions.

Example: To add 1 to each element, you need to use a loop or list comprehension:

my_list = [1, 2, 3]

my_list = [x + 1 for x in my_list]  # Using list comprehension

NumPy Arrays:

NumPy arrays support element-wise operations (like addition, multiplication, etc.) directly. You can apply mathematical operations to entire arrays without the need for explicit loops.

Example: Adding 1 to each element in a NumPy array:

import numpy as np

np_array = np.array([1, 2, 3])

np_array = np_array + 1  # This adds 1 to each element

Conclusion: NumPy arrays support vectorized operations (performing operations on entire arrays) directly, while Python lists require loops or comprehensions.

5. Dimensionality

Python Lists:

Python lists are essentially one-dimensional or nested lists, meaning that they can represent 2D or higher-dimensional data using lists of lists, but this is not as efficient or straightforward as NumPy.

Example: A 2D list in Python:

my_list = [[1, 2, 3], [4, 5, 6]]

NumPy Arrays:

NumPy arrays are multi-dimensional by design. You can easily create 1D, 2D, 3D, and even higher-dimensional arrays.

Example: Creating a 2D NumPy array:

np_array = np.array([[1, 2, 3], [4, 5, 6]])

Conclusion: NumPy arrays are designed for multi-dimensional data, whereas Python lists are limited to 1D or nested lists.

6. Indexing and Slicing

Python Lists:

Python lists support basic indexing and slicing, but operations on slices create copies of the data. This means that modifying a slice does not affect the original list.

NumPy Arrays:

NumPy arrays provide more advanced and efficient indexing and slicing features, including the ability to perform fancy indexing (using arrays or lists as indices), boolean indexing, and viewing slices (modifying a slice of a NumPy array affects the original array).

Example of slicing in NumPy:

np_array = np.array([1, 2, 3, 4, 5])

np_slice = np_array[1:4]  # Slicing from index 1 to 3

Conclusion: NumPy arrays offer more advanced and efficient indexing and
slicing capabilities compared to Python lists.

7. Shape and Reshaping

Python Lists:

Reshaping Python lists (like converting a 1D list to a 2D list) is not straightforward and typically requires manually modifying the structure.

NumPy Arrays:

NumPy provides built-in methods to easily reshape arrays (e.g., .reshape()) and perform operations across dimensions.

Example: Reshaping a 1D array to a 2D array:

np_array = np.array([1, 2, 3, 4, 5, 6])

np_array = np_array.reshape(2, 3)  # Reshapes to a 2x3 matrix

Conclusion: NumPy arrays have built-in methods for reshaping and handling multi-dimensional data, while Python lists do not.

8. Use Case

Python Lists:

Best used for general-purpose programming where flexibility and ease of use are more important than performance (e.g., handling mixed data types, lists of objects, etc.).

NumPy Arrays:

Best used for numerical computations, scientific computing, and handling large datasets where performance and memory efficiency are critical.

**Question No7:** What is a heatmap, and when should it be used?

**Answer:**

A heatmap is a data visualization technique that displays matrix-like data in a graphical format, where individual values are represented by colors. It provides an intuitive way to see patterns, correlations, or trends in data, especially when dealing with large datasets.

In a heatmap:

Color intensity or a specific color scale is used to represent numerical values in a two-dimensional space (such as a table or matrix).

Higher values can be represented by one color (e.g., red), while lower values are represented by another color (e.g., blue). A gradient of colors is typically used for this purpose.

A heatmap is often used when you have two variables (often categorical) that have a relationship, and you want to represent that relationship in a visually compact, easily interpretable manner.

Key Features of a Heatmap:

* Color Encoding: The primary feature of a heatmap is that it uses color to represent the magnitude of the data. The color gradient often corresponds to values, where darker colors or more intense colors represent higher values and lighter or different colors represent lower values.

* Matrix Layout: Heatmaps are typically displayed in a grid layout, where rows and columns represent data variables, and each cell’s color corresponds to the value of the data at the intersection of those rows and columns.

* Color Bar: A heatmap usually includes a color bar or legend that shows the scale of values mapped to colors.

When Should a Heatmap Be Used?

A heatmap is particularly useful when you want to:

1. Visualize Correlation Between Variables:

Heatmaps are often used in correlation matrices to show how different variables in a dataset are related to each other.
Example: You might use a heatmap to represent the correlation between features (columns) in a dataset, where darker colors indicate strong positive or negative correlations, and lighter colors indicate weak or no correlation.

2. Explore Patterns in Data:

Heatmaps are ideal for exploring patterns or trends in complex datasets, such as changes over time or spatial data.

Example: Visualizing hourly traffic data over a week or the intensity of sales in different geographic regions.

3. Show Relationships Between Categorical Variables:

Heatmaps can be used to represent relationships between categorical variables, especially when there are many categories and you want to see how they interact. You can use them to show the frequency of occurrences or any other summary statistic (e.g., mean, count).

Example: A heatmap can represent a contingency table showing how two categorical variables (e.g., Gender and Product Preference) interact.

4. Represent Large Data Sets:

Heatmaps are helpful when visualizing large amounts of data in a compact, readable way. For example, you can represent the values of a 2D matrix, where each element corresponds to a cell in the heatmap.

Example: A heatmap showing pixel intensities in image data or showing large biological datasets like gene expression levels across various conditions or samples.

5. Track Performance Metrics:

In business analytics, heatmaps can be used to display metrics across various dimensions, helping to identify high-performing areas or regions that need attention.

Example: A heatmap of sales performance in different regions or product categories.

6. Visualize Temporal Data:

Heatmaps can also be used to visualize temporal data, such as changes over days, months, or years.

Example: A time series heatmap might show how sales change over the days of the week across different months.



**Question No8:** What does the term “vectorized operation” mean in NumPy?

**Answer:**

In NumPy, the term "vectorized operation" refers to performing operations on entire arrays or matrices (vectors) in a way that avoids the need for explicit loops. Instead of writing a loop to iterate over each element of the array, vectorized operations allow you to apply a function or mathematical operation to the entire array at once. These operations are faster and more efficient because they leverage optimized C code under the hood, reducing the need for Python-level iteration.

Key Features of Vectorized Operations:

Efficiency: By using low-level optimizations (written in C), NumPy can execute operations on entire arrays without needing to iterate over them in Python, which is much slower.

Concise Code: Vectorized operations allow for cleaner, shorter, and more readable code. Instead of using loops, you can apply operations directly on arrays.

Broadcasting: NumPy can automatically handle arrays of different shapes during vectorized operations through broadcasting. It allows operations between arrays of different sizes by "stretching" smaller arrays to match the dimensions of larger ones.

Common Vectorized Operations in NumPy:

Element-wise operations (addition, multiplication, etc.): a + b, a * b

Mathematical functions: np.sin(a), np.log(a)

Array manipulation: np.sum(a), np.mean(a)

Advantages:

Speed: Vectorized operations are typically faster because they avoid the overhead of Python loops and use efficient low-level implementations.

Clarity: The code is usually simpler and more intuitive when using vectorized operations.

Thus, vectorization is a powerful feature in NumPy that helps to write concise and fast code when working with arrays.

Key Features and Significance of Hierarchical Indexing:

Multi-dimensional data representation:

Hierarchical indexing allows you to represent and manipulate data with multiple dimensions in a flat table.
Instead of having a single index for rows or columns, you have multiple levels of indexing, which can provide a richer structure to your data.

Group-based operations:

It is particularly useful for grouping data by multiple categories, making operations like aggregation, filtering, and transformation much more straightforward.
With hierarchical indexing, you can easily apply group-based operations without manually handling nested loops.

Improved data organization:

You can organize data in a more logical and natural way, where data is indexed according to multiple levels, such as by region and year, or department and employee.
For example, when working with time series data, you can index by date and category, or when analyzing financial data, you can index by company and quarter.

Accessing subsets of data:

Hierarchical indexing allows easy access to specific slices or subsets of data based on one or more index levels. You can access data using partial indexing, which simplifies working with subsets.
You can select data at any level of the hierarchy (e.g., selecting data for a specific year or category) without needing to manually filter or flatten the structure.

Efficient handling of missing data:

Hierarchical indexing can handle missing values in a way that maintains the integrity of the structure, especially when dealing with multi-dimensional datasets where some combinations of index levels may be absent.

Complex reshaping of data:

It facilitates the reshaping and pivoting of data. You can pivot or unpivot the data to represent it in different ways, which is useful for preparing data for analysis or reporting.
With hierarchical indexing, operations like stacking (converting columns into rows) and unstacking (converting rows into columns) are easier to perform.

It is especially useful for exploring pairwise relationships in multivariate data, providing a grid of scatterplots for each combination of features. It also helps you identify patterns, correlations, and distributions within the data.

Key Roles of pairplot():

Visualizing Pairwise Relationships:

The primary role of pairplot() is to create a grid of scatterplots that show the relationships between each pair of variables in a dataset. This allows you to quickly assess whether certain variables are correlated, if there are any linear or non-linear relationships, or if there are any outliers.

Diagonal Histograms/Kernel Density Estimation:

On the diagonal of the plot, pairplot() typically displays the distribution of each individual variable. By default, it shows histograms for numerical variables, but you can customize it to display other plots like Kernel Density Estimations (KDE).

Exploratory Data Analysis (EDA):

pairplot() is a great tool for performing exploratory data analysis (EDA), especially when you have a large dataset with multiple numerical features. It helps you quickly visualize the relationships, trends, and distributions in the data, guiding you toward useful insights.

Identifying Correlations:

By visualizing the relationships between pairs of variables, you can identify whether two variables exhibit a linear or non-linear correlation. Strong correlations (positive or negative) can often be visually identified by the scatterplot shapes (e.g., straight lines, curved lines).

Categorical Differentiation:

If your dataset includes categorical variables, you can color-code the points in the scatterplots using the hue parameter to represent different categories. This adds an extra layer of insight by helping you see how the categories (e.g., classes, groups) are distributed across the feature space.

Handling Large Datasets:

While pairplot() works well for smaller datasets, for large datasets, it can be slow and cluttered. However, it can be paired with other techniques, like subsampling the data, to handle larger datasets effectively.

Key Features of pairplot():

**Question No17:**  Explain the difference between apply() and map() in Pandas?

**Answer:**

In Pandas, both apply() and map() are used to apply functions to data, but they differ in how they operate and the types of objects they are used with. Here’s a breakdown of the key differences between apply() and map() in Pandas:

1. Usage with Data Structures

apply() is more general-purpose and can be used with both Series and DataFrames.
When used on a Series, it applies a function along the axis (to each element).
When used on a DataFrame, it can apply a function to either rows or columns (along a specified axis).

map() is specifically designed for Series (one-dimensional data). It is used to apply a function element-wise to each value in a Series. It is not directly applicable to DataFrames, although you can use it on each column individually within a DataFrame.

2. Function Types

apply(): The function passed to apply() can be any function that works on a single value (like a lambda function), or it can be a predefined function. When used on a DataFrame, you can apply more complex functions, like aggregations or custom functions that return a value for each row/column.

map(): The function passed to map() is typically element-wise. It works well for mapping values (e.g., from one value to another) and for applying simple transformations. It can also accept a dictionary, Series, or function to map values directly.

3. Return Type

apply(): When used on a Series, it returns a Series or a scalar depending on the function applied. When used on a DataFrame, it can return:
A Series if you apply the function to a single column.
A DataFrame if you apply the function to multiple columns or rows.

map(): Always returns a Series. When used on a Series, it returns a new Series with the transformed values. If you pass a dictionary or another Series to map(), the resulting Series will have the values replaced according to the dictionary/Series mapping.

4. Performance

map() is generally faster than apply() when working on Series because map() is optimized for element-wise mapping and directly uses a lookup dictionary or Series for transformation.

apply() is more flexible and powerful, but it may have a slightly slower performance compared to map() when applying simple transformations to individual elements.

5. Use Cases

apply():

It is often used for applying more complex operations, such as aggregations, transformations, or applying custom functions to rows/columns of a DataFrame.
You can use apply() to apply functions along rows or columns of a DataFrame (e.g., summing values across rows or applying a function to each column).

map():

It is best suited for simple element-wise operations like mapping values, handling missing data, or transforming data based on a dictionary or Series.

**Question No21:** Why is NumPy’s array slicing faster than Python’s list slicing?

**Answer:**

NumPy's array slicing is faster than Python's list slicing due to several key differences in how NumPy arrays and Python lists are implemented and how they handle memory and operations. Here are the main reasons for this speed difference:

1. Contiguous Memory Allocation:

NumPy Arrays: NumPy arrays are implemented as a contiguous block of memory. This means that the elements of a NumPy array are stored in a continuous, fixed-size memory location. When you slice a NumPy array, it can directly refer to the part of the memory that corresponds to the slice, without needing to copy the data. This results in faster slicing operations.
Python Lists: Python lists, on the other hand, are implemented as arrays of pointers to objects. These objects may be scattered throughout memory, leading to more overhead when slicing. In order to slice a list, Python needs to create a new list and copy the relevant elements into it, which involves more work than simply referencing part of a contiguous memory block.

2. No Copying of Data:

NumPy Slicing: When you slice a NumPy array, the slice typically does not
create a copy of the data. Instead, it returns a view of the original array. This means that NumPy can quickly create a reference to the selected portion of the array without allocating new memory. Any changes made to the slice will reflect in the original array (unless you explicitly copy it).
Python List Slicing: In contrast, Python’s list slicing creates a new list and copies the elements. The process of copying data involves more overhead and can be slower, especially for large lists.

3. Optimized Low-Level Implementation:

NumPy: NumPy is implemented in C and highly optimized for numerical operations. Its slicing mechanism is part of this low-level, performance-optimized code, which means it can perform operations like slicing much more efficiently.
Python Lists: Python lists are part of the standard Python implementation, which is written in C, but the slicing operation is not as optimized as NumPy’s slicing. List slicing in Python requires more Python-level checks and operations, making it slower in comparison.

4. Efficient Memory Usage:

NumPy Arrays: NumPy arrays are designed to store large amounts of data efficiently, and their internal structure is optimized for numerical computation. NumPy uses less memory overhead compared to Python lists, allowing slicing to be done faster due to more efficient memory management.
Python Lists: Python lists, by contrast, store elements as objects, each of which carries additional metadata (e.g., type, size, reference count). This leads to more memory overhead and slower performance when slicing.

5. Advanced Indexing and Stride Mechanisms:

NumPy: NumPy allows for more complex slicing, such as advanced indexing with strides, where you can slice arrays with custom step sizes, or even reverse slices. These operations are implemented in highly optimized C code that is fast and efficient.
Python Lists: Python lists support basic slicing with start, stop, and step arguments, but it does not offer the same level of optimization for advanced indexing or stride-based slicing.

**Question No22:**  What are some common use cases for Seaborn?

**Answer:**

Seaborn is a powerful data visualization library built on top of Matplotlib, which is specifically designed for creating informative and attractive statistical graphics. Seaborn makes it easier to create complex visualizations with just a few lines of code, providing a high-level interface for drawing attractive and informative statistical graphics.

Here are some common use cases for Seaborn:

1. Visualizing Distributions

Seaborn provides several functions to visualize the distribution of data, which is often the first step in understanding the underlying patterns of a dataset.

Histogram (sns.histplot()): Shows the distribution of a single continuous variable.
Kernel Density Estimate (KDE) plot (sns.kdeplot()): Plots a smooth estimate of the data's probability density function.
Box Plot (sns.boxplot()): Displays the summary statistics of a dataset, including median, quartiles, and potential outliers.
Violin Plot (sns.violinplot()): Combines aspects of box plots and KDE plots, giving a richer view of the data distribution.

2. Comparing Multiple Categories

Seaborn is great for comparing values across different categories using visual representations. Common types of plots for this purpose include:

Bar Plot (sns.barplot()): Displays the aggregated value of a continuous variable across categories.
Count Plot (sns.countplot()): Shows the number of occurrences of different categories.
Point Plot (sns.pointplot()): Displays estimates of central tendency (like means) with confidence intervals.
Facet Grid (sns.FacetGrid()): Used to create a grid of subplots for visualizing relationships between variables conditioned on categorical variables.

3. Visualizing Relationships Between Variables

Seaborn helps visualize relationships between variables using scatter plots and regression plots.

Scatter Plot (sns.scatterplot()): Plots a scatterplot to examine the relationship between two continuous variables.
Regression Plot (sns.regplot()): Fits a regression line to a scatterplot, helping to visualize trends and correlations between two variables.
Pair Plot (sns.pairplot()): Displays scatter plots for all pairwise combinations of variables in a DataFrame, useful for visualizing relationships in multi-dimensional datasets.

4. Visualizing Categorical Data with Statistical Aggregation

Seaborn allows you to visualize categorical data while aggregating the data using statistical methods. This is useful when working with large datasets where direct values for each category are not easily interpretable.

Bar Plots and Box Plots with Aggregated Data: You can aggregate data within categories and summarize it using mean, median, or other statistics.
Strip Plot (sns.stripplot()): A scatter plot for categorical data that plots individual data points and optionally uses jitter to avoid overlap.
Swarm Plot (sns.swarmplot()): Like a strip plot, but points are adjusted to avoid overlapping.

5. Correlation Heatmaps

Seaborn excels at creating heatmaps to visualize the correlation between different variables in a dataset.

Heatmap (sns.heatmap()): Creates a heatmap from a 2D matrix of values, often used to represent correlations between features in a dataset.

6. Time Series Data Visualization

Seaborn also supports time series visualizations, particularly when combined with Pandas’ datetime indexing capabilities.

Line Plot (sns.lineplot()): Shows trends over time or continuous variables, useful for time series data.


7. Visualizing Multivariate Data

Seaborn is useful for visualizing complex, multi-dimensional datasets. It can easily handle relationships between more than two variables using color, size, or other aesthetics.

Facet Grid: Allows for plotting different subplots based on additional categorical variables.
Heatmap with Clustered Data: Clustering data and plotting a heatmap to display patterns.

8. Customizing Visualizations

Seaborn also simplifies the process of customizing plots. It offers various aesthetic settings, such as color palettes, style adjustments, and plot themes to make visualizations more attractive and suitable for presentation.

Color Palettes: Seaborn provides built-in color palettes like "deep", "muted", "pastel", etc.
Themes: Seaborn allows easy switching between different themes (darkgrid, whitegrid, dark, white, ticks) to change the style of the plot.

It’s an excellent tool for statistical data exploration, making it easier for analysts and data scientists to uncover patterns and insights in their data quickly and effectively.

**Question No11:** Perform matrix multiplication using NumPy?

**Answer:**

```python
import numpy as np

# Create two matrices (2D arrays)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication using the @ operator
C = A @ B

# Alternatively, you can use np.dot() for matrix multiplication
# C = np.dot(A, B)

# Output the result
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResult of Matrix Multiplication (C = A @ B):")
print(C)
```
